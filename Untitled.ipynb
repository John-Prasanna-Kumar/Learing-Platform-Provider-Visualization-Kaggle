```python
random_sample=district['pp_total_raw'].dropna().sample(district['pp_total_raw'].isnull().sum(),random_state=0)
```

```python
def impute_nan(district, variable, random_sample):
    district[variable+"_random"]=district[variable]
    ##It will have the random sample to fill the na
    random_sample=district[variable].dropna().sample(district[variable].isnull().sum(),random_state=0)
    ##pandas need to have same index in order to merge the dataset
    random_sample.index=district[district[variable].isnull()].index
    district.loc[district[variable].isnull(),variable+'_random']=random_sample
```

```python
impute_nan(district, 'pp_total_raw', random_sample)
```

```python
#Cleaning features
district['pct_black/hispanic'] = district['pct_black/hispanic'].str.replace('[','', regex=True)
district['pct_free/reduced'] = district['pct_free/reduced'].str.replace('[','', regex=True)
district['county_connections_ratio'] = district['county_connections_ratio'].str.replace('[','', regex=True)
district['pp_total_raw_random'] = district['pp_total_raw_random'].str.replace('[','', regex=True)
```

```python
#Separating and creating new features from pct_black/hispanic
Black_hispanic = district['pct_black/hispanic'].str.split(',', n=1, expand =True)
district['pct_black'] = pd.to_numeric(Black_hispanic[0])
district['pct_hispanic'] = pd.to_numeric(Black_hispanic[1])
district['Black_&_Hispanic'] = (district['pct_black']+district['pct_hispanic'])/2
```

```python
#Separating and creating new features from pct_free/reduced
Free_reduced = district['pct_free/reduced'].str.split(',', n=1, expand =True)
district['pct_free'] = pd.to_numeric(Free_reduced[0])
district['pct_reduced'] = pd.to_numeric(Free_reduced[1])
district['Free_&_Reduced'] = (district['pct_free']+district['pct_reduced'])/2
```

```python
#Separating and creating new features from county_connections_ratio
county_connections = district['county_connections_ratio'].str.split(',', n=1, expand = True)
district['Country'] = pd.to_numeric(county_connections[0])
district['Connections'] = pd.to_numeric(county_connections[1])
district['CC_Ratio'] = (district['Country']+district['Connections'])/2
```

```python
#Separating and creating new features from pp_total_raw
Total_Raw = district['pp_total_raw_random'].str.split(',', n=1, expand = True)
district['pp_Total'] = pd.to_numeric(Total_Raw[0])
district['pp_Raw'] = pd.to_numeric(Total_Raw[1])
district['Total_&_Raw'] = (district['pp_Total']+district['pp_Raw'])/2
```

# Engagement Data Visualization

```python
import os
import glob
```

```python
path = r'C:\Users\91868\Desktop\Python Code\Kaggle-Price prediction\Kaggle-Learning Platform\engagement_data'
all_files = glob.glob(path + "/*.csv")
li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    fn = os.path.splitext(filename)
    df['district_id'] = os.path.basename(fn[0])
    li.append(df)

engagement_data = pd.concat(li)
engagement_data = engagement_data.reset_index(drop=True)
```

```python
engagement_data.shape
```

```python
engagement_data.isnull().sum()
```

```python
engagement_data.info()
```

```python
engagement_data.describe(include='all')
```

```python
#drop missing values
engagement_data.dropna(inplace=True)
```

```python
engagement_data.shape
```

```python
engagement_data.isnull().sum()
```

```python
engagement_data['time'] = pd.to_datetime(engagement_data['time'])
engagement_data['lp_id'] = engagement_data['lp_id'].astype(int)
engagement_data['district_id'] = engagement_data['district_id'].astype(int)
```

```python
engagement_data.head()
```

```python
# group by time without index
engagement_data_by_date1 = engagement_data.groupby(by = 'time', as_index = False).agg('mean')
```

```python
#Distribution of Percentage Access with time
figure = plt.figure(figsize=[15,8])
sns.lineplot(data=engagement_data_by_date1, x='time', y='pct_access', color = '#D96552')
plt.xlabel('Date', size=14)
plt.ylabel('Parcent Access', size=14)
plt.legend(labels = ['Percent Access'], loc = 'upper left', frameon = False)
plt.title('Percentage Access Trend', size=16)
plt.grid(b=True, axis='y')
plt.show()
```

```python
figure = plt.figure(figsize=[15,8])
sns.lineplot(data= engagement_data_by_date1, x='time', y='engagement_index', color = '#D96552')
plt.xlabel('Date', size=14)
plt.ylabel('Engagement_Index', size=14)
plt.legend(labels = ['Engagement_Index'], loc = 'upper left', frameon = False)
plt.title('Engagement Index Trend', size=16)
plt.grid(b=True, axis='y')
plt.show()
```

```python
from pandas.plotting import lag_plot
plt.figure(figsize=(15,8))
lag_plot(engagement_data_by_date1.engagement_index)
plt.show()
```

```python
from pandas.plotting import lag_plot
plt.figure(figsize=(15,8))
lag_plot(engagement_data_by_date1.pct_access)
plt.show()
```
